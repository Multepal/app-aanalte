```python
import pandas as np
from glob import glob
from lxml import etree as ET
# import xml.etree.ElementTree as ET
```

```python
files = sorted(glob("xom-*.xml"))
```

```python
ns = {'tei': "http://www.tei-c.org/ns/1.0"}
```

# Check if valid

```python
for file in files:
#     print(file)
    try:
        tree = ET.parse(file)
    except Exception as e:
        print(e)
```

# Generate XSLT Lines

```python
varlines = []
copylines = []
for file in files:
    
    idparts = file.split('.')[0].split('-')
    fn = idparts[3]
    sn = idparts[-1]
    fid = f"f{fn}s{sn}"
    
    varline = f"<xsl:variable name=\"{fid}\" select=\"document('xom-escolios-folio-{fn}-side-{sn}.xml')\"/>"
    varlines.append(varline)
    
    lbline = f"<lb id=\"f{fn}-s{sn}\" />"
    copylines.append(lbline)
    copyline = f"<xsl:copy-of select=\"${fid}//tei:text/tei:body/*\"/>"
    copylines.append(copyline)
```

```python
print('\n'.join(varlines))
```

```python
print('\n'.join(copylines))
```
